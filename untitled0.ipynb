```python
df_raw = df_raw[df_raw["Country Code"] == "NGA"].reset_index(drop=True)
```

```python
year_cols = [col for col in df_raw.columns if re.match(r"^\d{4}$", str(col))]

# Keep only descriptive columns + year columns to simplify dataset
df = df_raw[["Indicator Name", "Indicator Code"] + year_cols].copy()

print(df.shape)
df.head()

```

```python
df = df_raw[["Indicator Name", "Indicator Code"] + year_cols].copy()

print(df.shape)
df.head()
```

```python
# Function to compute the longest continuous run of non-missing values in a time series row
def longest_run_nonmissing(row):
    longest = 0
    current = 0
    for col in year_cols:
        if pd.notna(row[col]):  # If this year has a value
            current += 1        # Increase the current streak
            longest = max(longest, current)  # Update maximum if needed
        else:
            current = 0         # Missing value breaks the streak
    return longest

# Percentage coverage of available years
df["coverage_%"] = df[year_cols].notna().mean(axis=1) * 100

# First year with a value
df["first_year"] = df[year_cols].apply(lambda row: row.first_valid_index(), axis=1)

# Last year with a value
df["last_year"] = df[year_cols].apply(lambda row: row.last_valid_index(), axis=1)

# Longest uninterrupted run of available data
df["longest_run"] = df.apply(longest_run_nonmissing, axis=1)

df[['Indicator Name','coverage_%','first_year','last_year','longest_run']].head(10)
```

```python
# converting first and last year to numeric

df["first_year"] = pd.to_numeric(df["first_year"], errors="coerce")
df["last_year"] = pd.to_numeric(df["last_year"], errors="coerce")

# filtering indicators
candidate_indicators = df[
    (df["longest_run"] >= 25) &
    (df["last_year"] >= 2022)
].reset_index(drop=True)

candidate_indicators[['Indicator Name','Indicator Code','longest_run','coverage_%','first_year','last_year']]

```
