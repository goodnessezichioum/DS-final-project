```python
workforce_df["Date_Employed"] = pd.to_datetime(workforce_df["Date_Employed"])
workforce_df["Employment_Years"] = (pd.Timestamp.today() - workforce_df["Date_Employed"]).dt.days / 365

sns.histplot(workforce_df["Employment_Years"], bins=20, kde=True)
plt.title("Years Since Employment")
plt.xlabel("Years")
plt.ylabel("Count")
plt.show()
```
